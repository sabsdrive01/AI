## 1. What is Prompting?

### üîç Definition & Purpose
- Prompting is the method of instructing an AI model to perform a task.
- A well-designed prompt improves accuracy, clarity, and control over the model's output.

### üß© Prompt Formula
**Prompt = Instruction + Context + Task**

Example:
- *Instruction:* ‚ÄúSummarize the following text.‚Äù
- *Context:* Provide the text.
- *Task:* Specify format, length, constraints.

```
prompt = """
Instruction: Summarize the text below.

Context:
LLMs use pattern recognition to generate responses.
They do not think like humans.

Task: Provide a summary in 2 bullet points.
"""
```

## 2. Basic Prompting Techniques

### üü¶ Zero-Shot Prompting
- The model is given no examples.
- Used when the task is simple or well-known.

### üü© Few-Shot Prompting
- Provide examples to guide the model.
- Useful for formatting consistency or niche tasks.

### üìè Clear Instructions, Delimiters, and Examples
- Use separators like `---`, `<>`, or triple quotes.
- Remove ambiguity by specifying format and scope.

```
zero_shot_prompt = """
Translate to Hindi: "AI will transform the future."
"""
```

```
few_shot_prompt = """
Convert sentences into polite requests.

Example:
Sentence: Close the door.
Polite: Could you please close the door?

Example:
Sentence: Send me the file.
Polite: Could you please send me the file?

Now you try:
Sentence: Turn off the fan.
Polite:
"""
```

## 3. Prompt Structure Best Practices

### üé≠ Role Assignment
Example: ‚ÄúYou are a helpful assistant who explains concepts simply.‚Äù

### üß† Step-by-Step Reasoning (Chain-of-Thought)
- Encourage logical breakdown.
- Avoid one-shot answers for complex tasks.

### üóÇÔ∏è Specifying Output Format
- JSON
- Tables
- Bullet points
- Numbered lists

Clear formats reduce hallucinations and improve consistency.

```
prompt_structure_example = """
You are a helpful assistant.

Explain the concept below using step-by-step reasoning.
Output format: JSON with keys: explanation, example.

Concept: Why do airplanes fly?
"""
```
